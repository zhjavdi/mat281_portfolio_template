EL conjunto de datos se denomina `Automobile_data.csv`, el cual contiene información tal como: compañia, precio, kilometraje, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import numpy as np
import os
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","Automobile_data.csv")).set_index('index')
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Elimine los valores nulos (Nan)

```python
cols = df.columns.tolist() # lista con nombres de columnas
df_nonull = pd.DataFrame([]) # definimos un DataFrame vacio

for col in cols: # a cada columna que posee numeros le revisamos si tiene valores NaN o no
    if df[col].notnull().all()==False: # revisaremos solo las columnas con valores NaN
        if (df[col].dtype==float or df[col].dtype==int): # actuaremos sobre las columnas con datos numericos
            if len(df_nonull)>0: # si df_nonull tiene datos, lo vamos actualizando
                df_nonull = pd.DataFrame(df_nonull.fillna(df[col].mean()), columns=cols) 
            df_nonull = pd.DataFrame(df.fillna(df[col].mean()), columns=cols) # rellenamos los valores nulos con la media de la columna
        
df_nonull
```

2. Encuentra el nombre de la compañía de automóviles más cara

```python
companys = df_nonull['company'].unique().tolist() # lista de compañias
aux = 0 
company_aux = []
for enterprise in companys:
    val = df_nonull[df_nonull["company"]==enterprise]['price'].mean() # calculamos la media de precios de cada empresa
    if val>=aux:
        aux = max(aux,val)
        company_aux.append([enterprise,val])
print("La empresa mas cara es:\n"+str(pd.DataFrame(company_aux, columns=['company','price_mean']).max())) # entregamos la empresa mas cara
```

3. Imprimir todos los detalles de Toyota Cars

```python
df_nonull[df['company']=='toyota'].describe(include='all')
```

4. Cuente el total de automóviles por compañía

```python
group = df_nonull.groupby('company').count()[['body-style']] # hacemos un grupo entre compañias y contamos
group # imprimimos cualquier columna del grupo, ya que son todas iguales
```

5. Encuentra el coche con el precio más alto por compañía

```python
group2 = df_nonull.groupby('company')
group2['price'].max()
```

6. Encuentre el kilometraje promedio (**average-mileage**) de cada compañía automotriz

```python
group2['average-mileage'].mean()
```

7. Ordenar todos los autos por columna de precio (**price**)

```python
df_nonull.sort_values(by='price')
```

## Problema 02


Siguiendo la temática de los automóviles, resuelva los siguientes problemas:

#### a) Subproblema 01
A partir de los siguientes diccionarios:

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'],
              'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 
                'Price': [29995, 23600, 61500 , 58900]}
```

* Cree dos dataframes (**carsDf1** y **carsDf2**) según corresponda.
* Concatene ambos dataframes (**carsDf**) y añada una llave ["Germany", "Japan"], según corresponda.

```python
carsDf1 = pd.DataFrame(GermanCars)
carsDf1
```

```python
carsDf2 = pd.DataFrame(japaneseCars)
carsDf2
```

```python
conj = set(carsDf1.columns.tolist()).union(set((carsDf2.columns.tolist())))
conj
```

```python
carsDf = pd.merge(carsDf1, carsDf2, how='outer', on=list(conj))
carsDf
```
